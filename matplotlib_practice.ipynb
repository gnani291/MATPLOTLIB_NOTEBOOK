```python
import pandas as pd
import matplotlib.pyplot as plt

# Example Data
df = pd.DataFrame({
    'x': [1, 2, 3, 4, 5, 6],
    'y': [2, 4, 1, 3, 5, 7],
    'category': ['A', 'B', 'A', 'B', 'C', 'C']
})

# Define color and marker style for each category
colors = {'A': 'red', 'B': 'blue', 'C': 'green'}
markers = {'A': 'o', 'B': 's', 'C': '^'}  # o=Circle, s=Square, ^=Triangle

# Plot each category separately
for cat in df['category'].unique():
    subset = df[df['category'] == cat]
    plt.scatter(
        subset['x'],
        subset['y'],
        color=colors[cat],
        marker=markers[cat],
        label=cat,
        s=100  # marker size
    )

plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend(title="Category")
plt.title("Different Colors & Markers by Category")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------
# 1. Load and prepare data (Pandas)
# -------------------------
data = {
    "Size_sqft": [750, 800, 850, 900, 950, 1000, 1050],
    "Price_k":   [150, 160, 170, 180, 190, 200, 210]
}
df = pd.DataFrame(data)

# Features (X) and Target (y)
X = df["Size_sqft"].values.reshape(-1, 1)  # shape (7, 1)
y = df["Price_k"].values.reshape(-1, 1)    # shape (7, 1)

# Normalize data for better training
X = (X - X.mean()) / X.std()

# -------------------------
# 2. Model Initialization (NumPy)
# -------------------------
m = X.shape[0]   # number of samples
W = np.random.randn(1, 1)  # weight
b = np.random.randn(1)     # bias
learning_rate = 0.01

# -------------------------
# 3. Training loop (NumPy)
# -------------------------
loss_history = []
for epoch in range(200):
    # Forward pass: prediction
    y_pred = np.dot(X, W) + b

    # Compute Mean Squared Error (loss)
    loss = np.mean((y_pred - y) ** 2)
    loss_history.append(loss)

    # Gradients
    dW = (2/m) * np.dot(X.T, (y_pred - y))
    db = (2/m) * np.sum(y_pred - y)

    # Update parameters
    W -= learning_rate * dW
    b -= learning_rate * db

# -------------------------
# 4. Visualization (Matplotlib)
# -------------------------
plt.subplot(1, 2, 1)
plt.plot(loss_history)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss")

plt.subplot(1, 2, 2)
plt.scatter(df["Size_sqft"], df["Price_k"], color='blue', label="Data")
pred_line = ((df["Size_sqft"] - df["Size_sqft"].mean()) / df["Size_sqft"].std()) * W.item() + b.item()

plt.plot(df["Size_sqft"], pred_line, color='red', label="Prediction")
plt.xlabel("Size (sqft)")
plt.ylabel("Price (k)")
plt.title("Model Fit")
plt.legend()

plt.tight_layout()
plt.show()

# -------------------------
# 5. Output
# -------------------------
print(f"Trained Weight: {W[0][0]:.4f}, Bias: {b[0]:.4f}")
```

```python
import matplotlib.pyplot as plt

# Data (you can adjust percentages if needed)
labels = ["MSMEs", "Large Companies", "Staffing Agencies & Gig Platforms"]
sizes = [40, 35, 25]   # assumed distribution
colors = ["#1f77b4", "#ff7f0e", "#2ca02c"]

# Create figure
fig, ax = plt.subplots(figsize=(7,7), dpi=160)

# Modern pie chart
wedges, texts, autotexts = ax.pie(
    sizes,
    colors=colors,
    startangle=90,
    autopct='%1.0f%%',
    pctdistance=0.7
)

# Customize text styles
for text in texts:
    text.set_fontsize(12)
for autotext in autotexts:
    autotext.set_fontsize(11)
    autotext.set_color("white")

# Title
ax.set_title("HR Tech Market Demand Distribution", fontsize=16, pad=20)

# Add legend
ax.legend(wedges, labels, title="Segments", loc="upper center",
          bbox_to_anchor=(0.5, -0.05), ncol=3)

plt.tight_layout()
plt.show()
```
